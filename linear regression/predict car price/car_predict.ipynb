```python
import pandas as pd
df = pd.read_excel("carprice.xlsx")
df
```

```python
df.fillna(method='ffill',inplace=True)
df
```

```python
df.rename(columns={
    'Car Model' :'car',
    'Mileage':'mileage',
    'Sell Price($)' : 'sell_price',
    'Age(yrs)':'age'
},inplace=True)
df
```

```python
dummies = pd.get_dummies(df.car)
dummies
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
final = merged.drop(['car','Mercedez'],axis="columns")
final
```

```python
X = final.drop(['sell_price'],axis="columns")
y = final.sell_price
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X,y)
reg
```

```python
#Predict price of a mercedez benz that is 4 years old with mileage 45000
#mileage , mercedez benz = 0,0
#mileage , BMW = 0, 1
#mileage , Audi = 1,0
reg.predict([[45000,4,0,0]])
```

```python
#Predict price of a BMW that is 7 years old with mileage 86000
reg.predict([[86000,7,0,1]])
```

```python
#accuracy model
acc = reg.score(X,y)
acc
```
