```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import math
```

```python
def initialze_zero(val):
    if val == math.isnan(val):
        return 'zero'
    return val
```

```python
df = pd.read_excel("datafile.xlsx") #pip install xlrd for support .xlsx file
df
```

```python
df.experience.fillna('zero',inplace=True)
df
```

```python
from word2number import w2n

exp =[]

for i in range(df.experience.shape[0]):
    exp.append(w2n.word_to_num(df.experience[i]))
    
del df['experience']

df
```

```python
df['exprience'] = exp
df
```

```python
df.fillna(6,inplace=True)#assume
df.rename(columns={"test_score(out of 10)" : "test","interview_score(out of 10)":"interview"},inplace=True)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['test','interview','exprience']],df['salary($)'])
```

```python
reg.predict([[9,10,13]])
```
