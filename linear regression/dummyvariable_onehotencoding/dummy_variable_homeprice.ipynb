```python
import pandas as pd

df = pd.read_excel("homeprice.xlsx")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town'],axis="columns") #remove town column
final
```

```python
#we have to remove 1 column in dummies columns 
final = final.drop(['surat'],axis="columns")
final
```

```python
X = final.drop(['price'],axis="columns")
X
```

```python
y = final.price
y
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg
```

```python
reg.fit(X,y)
```

```python
# area, 0,0 = surat
#area , 0,1 = goa
#area , 1,0 = bardoli
reg.predict([[4400,0,0]])
```
