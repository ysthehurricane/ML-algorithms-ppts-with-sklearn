```python
import pandas as pd
df = pd.read_csv('salaries.csv')
df
```

```python
inputs = df.drop('salary_more_then_100k',axis="columns")
inputs
```

```python
from sklearn.preprocessing import LabelEncoder

label_company = LabelEncoder()
label_job = LabelEncoder()
label_degree = LabelEncoder()
```

```python
inputs['cmny'] = label_company.fit_transform(inputs['company'])
inputs['jb'] = label_job.fit_transform(inputs['job'])
inputs['dg'] = label_degree.fit_transform(inputs['degree'])

inputs
```

```python
final =inputs.drop(['company','job','degree'],axis="columns")
```

```python
final.rename(columns={
    'cmny':'company',
    'jb':'job',
    'dg':'degree'
})
final
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(final,df['salary_more_then_100k'],test_size=0.3)
```

```python
reg = tree.DecisionTreeClassifier(criterion='entropy')
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
y_test
```

```python
reg.score(x_test,y_test)
```
